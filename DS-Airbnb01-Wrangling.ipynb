```python
#To manually update neighbourhoods that didn't fit into boroughs
neighborhood_borough_mapping = {
    'Mimico (includes Humber Bay Shores)': 'Etobicoke',
    'Wexford/Maryvale': 'Scarborough',
    'Cabbagetown-South St.James Town': 'Downtown Toronto',
    'North St.James Town': 'Downtown Toronto',
    "O'Connor-Parkview": 'East York',  
    'Weston-Pellam Park': 'York',
    'Yonge-Eglinton': 'Midtown Toronto',
    'Danforth East York': 'East York',
    'St.Andrew-Windfields': 'North York',
    'Yonge-St.Clair': 'Midtown Toronto',
    'Taylor-Massey': 'Scarborough'
}

# Iterate over the DataFrame rows where borough is null
for index, row in new_df[new_df['borough'].isnull()].iterrows():
    neighborhood = row['neighbourhood_cleansed']
    if neighborhood in neighborhood_borough_mapping:
        # Assign the correct borough based on the mapping dictionary
        new_df.loc[index, 'borough'] = neighborhood_borough_mapping[neighborhood]
```
